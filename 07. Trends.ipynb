```python
import pandas as pd
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
```

```python
subreddit_name = "Nietzsche"

df_subreddit = pd.read_csv(f"{subreddit_name}/{subreddit_name}_mentalhealth_common_author_posts_processed.csv")
df_subreddit['isodate'] = pd.to_datetime(df_subreddit["isodate"])
df_subreddit.drop(['index'], axis = 1, inplace = True)
df_subreddit.head()
```

```python
df_subreddit.groupby(by=['subreddit']).size().sort_values(ascending= False).head(20)
```

```python
type(df_subreddit['isodate'][0])
```

```python
def get_aggregated_df(agg_cols):
    groupby_str = f"df_subreddit.groupby(by =[pd.Grouper(key='isodate', axis=0, freq='D')] ).agg("
    for col in agg_cols:
        col_string = f"{col}=('{col}', np.sum),"
        groupby_str += col_string
    
    groupby_str = groupby_str[:-1]
    groupby_str += ").reset_index()"
    return eval(groupby_str)

aggregated_df = get_aggregated_df(df_subreddit.columns[-11:])
aggregated_df.head(40)
```
